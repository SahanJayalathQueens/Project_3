```python


# Constants
annual_interest_rate = 6.71
monthly_interest_rate = (annual_interest_rate / 100) / 12
loan_term_months = 30 * 12

# Function to calculate monthly payment
def calculate_monthly_payment(price):
    if pd.isna(price) or price <= 0:
        return 0  # Handle cases where the price is NaN or less than or equal to 0
    return price * (monthly_interest_rate * (1 + monthly_interest_rate) ** loan_term_months) / ((1 + monthly_interest_rate) ** loan_term_months - 1)

# Create a new column for monthly payment
df['monthly_payment'] = df['price'].apply(calculate_monthly_payment)

# Display the DataFrame with the new column
print(df[['price', 'monthly_payment']].head())
```

```python
import pandas as pd

# Load your DataFrame
df = pd.read_csv('your_file.csv')

# Constants
annual_interest_rate = 6.71
monthly_interest_rate = (annual_interest_rate / 100) / 12
loan_term_months = 30 * 12

# Display the DataFrame shape to confirm loading
print(df.shape)
```

```python
chunk_size = 1000  # Adjust this size based on your system's capacity
for i in range(0, df.shape[0], chunk_size):
    end = min(i + chunk_size, df.shape[0])
    chunk = df.iloc[i:end]
    
    # Calculate payments for valid prices in the chunk
    chunk.loc[chunk['price'] > 0, 'monthly_payment'] = (
        chunk.loc[chunk['price'] > 0, 'price'] *
        (monthly_interest_rate * (1 + monthly_interest_rate) ** loan_term_months) /
        ((1 + monthly_interest_rate) ** loan_term_months - 1)
    )
    
    # Update the main DataFrame
    df.loc[i:end, 'monthly_payment'] = chunk['monthly_payment']
    
    print(f"Processed rows {i} to {end}")

# Display the DataFrame with the new column
print(df[['price', 'monthly_payment']].head())
```
