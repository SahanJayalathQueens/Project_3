```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix

df = pd.read_csv(r'C:\Users\Sahan\Documents\Project_3\CSV\test.csv')
df.info()
```

```python
# Step 1: Data Cleaning
# Drop the Average Rent and street_name columns
df.drop(columns=['Average Rent', 'street_name'], inplace=True)

# Display the updated DataFrame info to confirm changes
print(df.info())
```

```python
# Constants
annual_interest_rate = 0.0673  # NYC mortgage rate
monthly_interest_rate = annual_interest_rate / 12
loan_term_years = 30
n_payments = loan_term_years * 12  # Total number of payments
taxes_and_fees = 700  # Additional monthly costs

# Step 1: Calculate Monthly Mortgage Payment
df['Monthly_Mortgage'] = (
    df['price'] * (monthly_interest_rate * (1 + monthly_interest_rate)**n_payments) / 
    ((1 + monthly_interest_rate)**n_payments - 1)
) + taxes_and_fees

# Check the updated DataFrame
print(df[['price', 'Monthly_Mortgage']].head())
```

```python
# Step: Drop the Mortgage column
df.drop(columns=['Mortgage'], inplace=True)

# Display the updated DataFrame info to confirm changes
print(df.info())
```

```python
df
```

```python

affordability_threshold = 0.30  # 30% of income

# Create Affordability Indicator
df['can_afford'] = (df['Monthly_Mortgage'] <= (df['Average Income'] / 12 * affordability_threshold)).astype(int)

# Check the updated DataFrame
print(df[['Average Income', 'Monthly_Mortgage', 'can_afford']].head())
```

```python
# Step: One-hot encode categorical variables without dropping the first category
df = pd.get_dummies(df, columns=['Borough', 'property_type'], drop_first=False)

# Check the updated DataFrame
print(df.head())
```

```python
# Compute the correlation matrix
correlation_matrix = df.corr()

# Display the correlation matrix
print(correlation_matrix)

plt.figure(figsize=(14, 20))

# Create a heatmap
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', square=True)

# Title and labels
plt.title('Correlation Matrix')
plt.show()
```

```python
# Step: Drop the price column
df.drop(columns=['price'], inplace=True)

# Check the updated DataFrame
print(df.info())
```

```python
# Step: Drop the bedroom_number column
df.drop(columns=['bedroom_number'], inplace=True)

# Check the updated DataFrame
print(df.info())
```

```python
# Define affordability threshold
affordability_threshold = 0.40

# Create the target variable
df['can_afford'] = (df['Monthly_Mortgage'] <= (df['Average Income'] / 12 * affordability_threshold)).astype(int)

# Check the updated DataFrame
print(df['can_afford'].value_counts())
```

```python
# Define features and target variable
X = df.drop(columns=['can_afford'])
y = df['can_afford']

# Perform train-test split
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Check the shapes of the splits
print(f"Training set shape: {X_train.shape}, Test set shape: {X_test.shape}")
```

```python
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Fit the scaler on the training data and transform both the training and test data
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Check the shapes after scaling
print(f"Scaled training set shape: {X_train_scaled.shape}, Scaled test set shape: {X_test_scaled.shape}")
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

# Initialize the model
logreg = LogisticRegression()

# Train the model
logreg.fit(X_train_scaled, y_train)

# Make predictions on the test set
y_pred = logreg.predict(X_test_scaled)

# Print the confusion matrix and classification report
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# Predict probabilities
y_prob = logreg.predict_proba(X_test_scaled)[:, 1]

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)

# Calculate AUC
auc = roc_auc_score(y_test, y_prob)

# Plot ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='blue', label=f'ROC Curve (AUC = {auc:.2f})')
plt.plot([0, 1], [0, 1], color='red', linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend(loc='lower right')
plt.show()
```

```python
from sklearn.model_selection import cross_val_score

# Perform cross-validation
cv_scores = cross_val_score(logreg, X_train_scaled, y_train, cv=5, scoring='roc_auc')

# Print cross-validation scores
print(f"Cross-Validation AUC Scores: {cv_scores}")
print(f"Mean AUC: {cv_scores.mean():.4f}, Std Dev: {cv_scores.std():.4f}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report

# Assuming df is your DataFrame with the features and target variable
X = df.drop('can_afford', axis=1)  # Features
y = df['can_afford']                # Target variable

# Split the data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Standardize the features
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

# Train the logistic regression model
model = LogisticRegression(C=100, penalty='l1', solver='liblinear', max_iter=500)
model.fit(X_train_scaled, y_train)

# Generate predictions
y_pred = model.predict(X_test_scaled)

# Evaluate the model
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print("Confusion Matrix:\n", conf_matrix)
print("\nClassification Report:\n", class_report)
```
